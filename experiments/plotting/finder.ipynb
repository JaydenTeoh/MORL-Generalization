```python
import pandas as pd

# Function to load CSV and calculate max values for each column
def calculate_max_values(csv_file_path):
    # Load the CSV file into a DataFrame
    df = pd.read_csv(csv_file_path)
    
    # Iterate over each column and print the maximum value
    for column in df.columns:
        if pd.api.types.is_numeric_dtype(df[column]):  # Only calculate max for numeric columns
            max_value = df[column].max()
            min_value = df[column].min()
            print(f"Column: {column}, Max value: {max_value}, Min value: {min_value}")
        else:
            print(f"Column: {column} is not numeric, skipping.")

# Example usage
csv_file_path = 'data/single_env/MOHumanoidDR-v5/MOHumanoidHard-v5.csv'  # Replace with your CSV file path
calculate_max_values(csv_file_path)
```
